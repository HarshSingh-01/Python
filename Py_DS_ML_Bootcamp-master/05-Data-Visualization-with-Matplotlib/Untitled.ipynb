```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,"r")
plt.xlabel("Time")
plt.ylabel("Speed")
plt.title("Speed vs Time Graph")
```

```python
plt.subplot(1,2,1)
plt.subplot(1,2,2)
plt.plot(x,y,'r--')
plt.plot(x,y,'b--')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(x,y,'b--')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.2,0.2,0.9,0.9])
axes.plot(x,y,'g*-')
axes.set_xlabel("Time")
axes.set_ylabel("Speed")
axes.set_title("Speed vs Time")
```

```python
# Creating a canvas
fig = plt.figure()

#Creating axes
axes1 = fig.add_axes([0.2,0.2,0.9,0.9])
axes2 = fig.add_axes([0.3,0.6,0.3,0.3])

axes1.plot(x,y,'g*-')
axes1.set_xlabel("Time")
axes1.set_ylabel("Speed")
axes1.set_title("Speed vs Time")

axes2.plot(y,x,'r--')
axes2.set_xlabel("Velocity")
axes2.set_ylabel("Time")
axes2.set_title("Velocity vs Time")
```

```python
fig,axes = plt.subplots()

axes.plot(x,y,'r--')
axes.set_xlabel("Time")
axes.set_ylabel("Speed")
axes.set_title("Speed vs Time")          
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# Display the figure object    
#fig
```

```python
fig = plt.figure(figsize=(12,4),dpi=100)
```

```python
fig = plt.figure(figsize=(8,4))

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label='x**2')
axes.plot(x,x**3,label="x**3")
axes.set_xlim(0,4)
plt.legend()
```

```python
fig, axes = plt.subplots(figsize=(8,4))

axes.plot(x,x+1, 'r--',alpha=0.5)
axes.plot(x,x+2, 'r+-',alpha=0.8)
axes.plot(x,x+3, 'r*-',alpha=1.0)
```

```python
fig,axes = plt.subplots(1,3,figsize=(12,4))

axes[0].plot(x,x**2, x,x**3)
axes[0].set_title("Plot1")

axes[1].plot(x,x**2, x,x**3)
axes[1].set_title("Plot2")
axes[1].axis('tight')

axes[2].plot(x,x**2, x,x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,4])
axes[2].set_title("Plot3")
```
